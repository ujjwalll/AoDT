```
import pandas as pd
df = pd.read_json('finalfile.json', encoding='utf-8', lines=True)
```

```
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="MIDAS")
from geopy.exc import GeocoderTimedOut
URL_INIT = 'https://twitter.com/'



list_of_users = unique_users

list_of_found_userlocations = []

list_of_nonfound_userlocations = []

def parse_url(tweet_user):
    url = URL_INIT+ tweet_user.strip('@')
    return url

for user in list_of_users:
    try:
        url = parse_url(user)
        response = urlopen(url)
    except:
        continue
    html = response.read()
    soup = BeautifulSoup(html)
    location = soup.find('span','ProfileHeaderCard-locationText')
    location = location.text.encode('utf8')
    location = location.splitlines()
    print("yay")
    list_of_found_userlocations.append(location)
```
